Attrition Analysis

```python
df['Attrition'].value_counts()
```

Age Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
```

Department Counts

```python
dep_count = df['Department'].value_counts()
sns.barplot(x=dep_count.index,y=dep_count.values,palette='rainbow')
```

Attrition By Department

```python
pd.crosstab(df['Department'],df['Attrition'], normalize='index') * 100
```

```python
sns.countplot(x='Department',hue='Attrition',data=df)
plt.title('Attrition By Department')
```

Attrition By Over Time

```python
pd.crosstab(['OverTime'],df['Attrition'],normalize='index') * 100
```

```python
sns.countplot(x='OverTime',hue='Attrition',data=df,palette='rainbow')
plt.title('Attrition By OverTime')
```

Attrition By Job Role

```python
pd.crosstab(['JobRole'],df['Attrition'],normalize='index') * 100
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole',hue='Attrition',data=df)
plt.xticks(rotation=45)
plt.title('Attrition By Job Role')
```

Attrition Rate Calculation

```python
attr_rate = df['Attrition'].value_counts(normalize=True) * 100
attr_rate
```
